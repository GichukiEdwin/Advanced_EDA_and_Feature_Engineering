```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
max_temp = [32, 34, 31, 33, 35, 36, 33]
min_temp = [20, 22, 21, 23, 24, 25, 22]
avg_temp = [26, 28, 26, 28, 29, 30, 27]
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, max_temp,label='max')
plt.plot(days, avg_temp,label='avg')
plt.plot(days, min_temp,label='min')
plt.legend(loc='best',shadow=True)
plt.grid()
```

```python
# List of tech company abbreviations
companies = ['AAPL', 'MSFT', 'AMZN', 'GOOGL', 'META', 'TSLA', 'NVDA', 'ORCL', 'IBM', 'INTC']
# Revenues (in billions of USD)
revenues = [394.3, 211.9, 524.9, 283.0, 116.6, 81.5, 54.3, 50.0, 60.5, 63.1]
# Profits (Net Income in billions of USD)
profits = [99.8, 72.7, 21.4, 59.9, 29.1, 12.6, 16.6, 8.5, 1.7, 8.0]
# Market Capitalization (in billions of USD)
market_caps = [2725, 2441, 1360, 1856, 791, 814, 1040, 314, 140, 149]
# Earnings Per Share (EPS)
eps = [6.15, 9.68, 2.19, 4.56, 10.42, 3.62, 8.54, 3.43, 1.95, 1.89]
# Price-to-Earnings (P/E) Ratio
pe_ratios = [29, 33, 64, 31, 26, 64, 31, 22, 73, 30]
# Return on Equity (ROE) as a percentage
roe = [147.4, 42.8, 10.3, 30.5, 23.2, 19.4, 40.4, 18.3, 7.4, 17.1]
xpos=np.arange(len(companies))
xpos
```

```python
plt.figure(figsize=(15,10))
plt.xticks(xpos,companies)
plt.title('US Tech Stocks')
bar_width = 0.15
plt.bar(xpos - 2*bar_width, revenues, width=bar_width, label='Revenue')
plt.bar(xpos - bar_width, profits, width=bar_width, label='Net Income')
plt.bar(xpos, market_caps, width=bar_width, label='Market Capitalization')
plt.bar(xpos + bar_width, eps, width=bar_width, label='Earnings per share')
plt.bar(xpos + 2*bar_width, roe, width=bar_width, label='Returns on equity')

plt.legend()
```

```python
# Example blood sugar levels (in mg/dL) for men and women
blood_sugar_men = [85, 90, 110, 115, 120, 130, 95, 100, 105, 160, 170, 180, 140, 135, 125, 145, 150, 155, 200, 190]
blood_sugar_women = [80, 88, 92, 102, 112, 118, 125, 132, 95, 105, 108, 160, 150, 140, 170, 155, 120, 130, 100, 110]
plt.figure(figsize=(10, 6))
plt.xlabel('Blood Sugar Range (mm/dl)')
plt.ylabel('Number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=3,rwidth=0.95,label=['Men','Women'], color=['aqua','pink'])
plt.legend()
```
