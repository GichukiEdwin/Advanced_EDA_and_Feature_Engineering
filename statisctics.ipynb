```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd

# The dataset
data = [12, 13, 14, 14, 15, 15, 15, 16, 16, 16, 16, 
        17, 17, 17, 17, 18, 18, 18, 19, 19, 20, 20, 21, 22, 23]

# Convert to numpy array and pandas Series for convenience
data_array = np.array(data)
data_series = pd.Series(data)

print("=== Summary Statistics ===")
print(f"Count (n)     : {len(data)}")
print(f"Mean          : {np.mean(data_array):.2f}")
print(f"Median        : {np.median(data_array):.1f}")
print(f"Mode          : {stats.mode(data_array).mode}")
print(f"Std Dev       : {np.std(data_array, ddof=1):.2f}")
print(f"Min           : {np.min(data_array)}")
print(f"Max           : {np.max(data_array)}")
print(f"Q1 (25%)      : {np.percentile(data_array, 25):.1f}")
print(f"Q3 (75%)      : {np.percentile(data_array, 75):.1f}")

# ====================== Visualizations ======================

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

fig = plt.figure(figsize=(12, 8))

# 1. Boxplot
plt.subplot(2, 2, 1)
sns.boxplot(y=data_array, color='skyblue')
plt.title('Boxplot - Symmetrical Distribution')
plt.ylabel('Value')

# 2. Histogram with KDE (shows unimodal & symmetrical curve)
plt.subplot(2, 2, 2)
sns.histplot(data_array, kde=True, bins=12, color='skyblue', edgecolor='black')
plt.title('Histogram + KDE\n(Unimodal & Symmetrical)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 3. Density plot (smooth curve)
plt.subplot(2, 2, 3)
sns.kdeplot(data_array, fill=True, color='skyblue')
plt.title('Density Plot (Symmetrical Bell Curve)')
plt.xlabel('Value')

# 4. Strip plot + boxplot (shows individual points)
plt.subplot(2, 2, 4)
sns.boxplot(y=data_array, color='lightblue')
sns.stripplot(y=data_array, color='red', alpha=0.7, size=8, jitter=True)
plt.title('Boxplot with Individual Data Points')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(y=data_array, ax=axes[0])
sns.histplot(data_array, kde=True, bins=12, ax=axes[1])
plt.suptitle('Symmetrical Distribution - Boxplot & Histogram')
plt.show()
```
