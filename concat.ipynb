```python
import pandas as pd
```

```python
et_cities=pd.DataFrame({
    'cities': ['Dire Dawa', 'Makelle', 'Adama', 'Harar', 'Nekemte'],
    'temperature': [21, 25, 33, 53, 49],
    'humidity': [78, 87, 94, 89, 73]
})
et_cities
```

```python
so_cities=pd.DataFrame({
    'cities': ['Hargeisa', 'Bossaso', 'Garowe', 'Mogadishu', 'Baidoa'],
    'temperature': [50, 47, 40, 56, 43],
    'humidity': [73, 80, 97, 99, 93]
})
so_cities
```

```python
df=pd.concat([et_cities, so_cities])
df
```

```python
df_1=pd.concat([et_cities,so_cities],ignore_index=True)
df_1
```

```python
df_2=pd.concat([et_cities,so_cities],keys=['Ethiopia', 'Somalia'])
df_2
```
