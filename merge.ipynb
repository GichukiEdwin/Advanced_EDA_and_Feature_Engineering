```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Mumbai', 'Boston', 'Moscow', 'San Francisco'],
    'temperature': [27, 18, 34, 39, 49, 32]
})
df.head(10)
```

```python
df_1 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Mumbai', 'Boston', 'Moscow', 'Baltimore'],
    'humidity': [78, 97, 64, 89, 91, 67]
})
df_1.head(10)
```

```python
df_2 = pd.merge(df, df_1, on='city')
df_2.head(10)
```

```python
df_3 = pd.merge(df, df_1, on = 'city', how = 'outer')
df_3.head(10)
```

```python
df_4 = pd.merge(df, df_1, on = 'city', how = 'outer', indicator = True)
df_4.head(10)
```

```python
df_5 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Mumbai', 'Boston', 'Moscow', 'Baltimore'],
    'temperature': [19, 12, 40, 29, 22, 39],
    'humidity': [78, 97, 64, 89, 91, 67]
})
df_5.head(10)
```

```python
df_6 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Mumbai', 'Boston', 'Moscow', 'San Francisco'],
    'temperature': [27, 18, 34, 39, 49, 32],
    'humidity': [69, 80, 74, 99, 73, 87]
})
df_6.head(10)
```

```python
df_7 =  pd.merge(df_5, df_6, on = 'city')
df_7.head(10)
```

```python
df_8 = pd.merge(df_5, df_6, on = 'city', suffixes=['_left', '_right'])
df_8.head(10)
```
